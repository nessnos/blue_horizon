```python
import pandas as pd
from sqlalchemy import create_engine

engine = create_engine('postgresql://postgres:waterize@localhost:5422/waterdata')

chunk_size = 500000
metadata_columns = [
    'metadata_versionId', 
    'metadata_beginLifeSpanVersion', 
    'metadata_statusCode', 
    'metadata_observationStatus', 
    'metadata_statements'
]

chunks = pd.read_csv('../data/AggregatedWaterData.csv', chunksize=chunk_size)

for chunk in chunks:
    chunk = chunk.drop(columns=metadata_columns)

    int_columns = ["phenomenonTimeReferenceYear", "resultQualityNumberOfSamplesBelowLOQ", "resultQualityMinimumBelowLOQ", "resultQualityMeanBelowLOQ", "resultQualityMaximumBelowLOQ", "resultQualityMedianBelowLOQ"]
    for col in int_columns:
        chunk[col] = pd.to_numeric(chunk[col], errors='coerce').fillna(0).astype("Int64")

    float_columns = ["procedureLOQValue", "resultNumberOfSamples", "resultMinimumValue", "resultMeanValue", "resultMaximumValue", "resultMedianValue", "resultStandardDeviationValue", "parameterSampleDepth"]
    for col in float_columns:
        chunk[col] = pd.to_numeric(chunk[col], errors='coerce').fillna(0)

    chunk.to_sql('startapp_waterdata', engine, if_exists='append', index=False)

    print(f"Processed and inserted chunk {chunk.head()}")
```
